# 일반인 이미지 제거
각 7개의 감정(기쁨, 불안, 당황, 중립, 상처, 슬픔, 분노)별 전문인/일반인 이미지 중에서  
데이터의 퀄리티를 향상시키고자 일반인 이미지를 제거한다.

## 기쁨

'기쁨' 라벨 데이터를 받아와 DataFrame의 형태로 변환합니다.  
사용할 전문인 이미지의 개수를 확인한다.

```python
filename = "img_emotion_training_data(기쁨).json"
happy_label_df = pd.read_json(filename)
print(filename, happy_label_df.shape)
display(happy_label_df.head(10))

happy_label_df[happy_label_df.isProf == "전문인"].count()
```

한글 이슈를 해결하기 위해서 별도로 'hash_value' column을 만듭니다.  

```python
filenames = happy_label_df["filename"].tolist()
happy_label_df["hash_value"] = [filename.split("_")[0] for filename in filenames]
```

일반인에 해당하는 row들을 추려서 list형태로 변환합니다.

```python
general_label_list = happy_label_df[happy_label_df.isProf == "일반인"].hash_value.tolist()
general_label_list[:10]
```

```python
len(general_label_list)
```

각 감정에 해당하는 이미지 폴더에 접근하여 list 형태로 반환합니다.  
한글 이슈를 해결하기 위해서 파일명에서 제일 앞 부분인 해시값을 기준으로  
앞에서 추려낸 일반인 list에 존재하는지 유무를 판별하여 제거할 일반인 이미지 리스트를 만듭니다.  

```python
general_image_list = []
path = '[원천]EMOIMG_기쁨_TRAIN_01'
image_list = os.listdir(path)
happy_image_list = [image.split("_") for image in image_list if image.endswith('.jpg')]

for happy_image in happy_image_list:
  if happy_image[0] in general_label_list:
    general_image_list.append('_'.join(happy_image))
else:
  print(f'{len(happy_image_list)}의 이미지 중에서 {len(general_image_list)}개의 일반인 사진을 검출하였습니다.')
```

일반이 이미지 리스트 파일명을 이용하여 해당 폴더에서 제거합니다.

```python
for f_name in general_image_list:
  os.remove(f'[원천]EMOIMG_기쁨_TRAIN_01/{f_name}')
```

폴더에서 일반인 이미지가 제거된 것을 확인합니다.

```python
general_image_list = []
path = '[원천]EMOIMG_기쁨_TRAIN_01'
image_list = os.listdir(path)
happy_image_list = [image.split("_") for image in image_list if image.endswith('.jpg')]

for happy_image in happy_image_list:
  if happy_image[0] in general_label_list:
    general_image_list.append('_'.join(happy_image))
else:
  print(f'{len(happy_image_list)}의 이미지 중에서 {len(general_image_list)}개의 일반인 사진을 검출하였습니다.')
```
